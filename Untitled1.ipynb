```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
companies = ['AAPL','LLY','MSFT','TSLA','AMZN','GOOGL','JPM','META','NVDA','TSM']
start = '2013-01-01'
end = '2023-01-01'
data = yf.download(companies,start,end)
data
```

```python
data=data['Adj Close']
data.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
#ho: Non-stationary
#h1: Stationary
def adfuller_test(e):
    result=adfuller(e)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
adfuller_test(data['AAPL'])
adfuller_test(data['LLY'])
adfuller_test(data['MSFT'])
adfuller_test(data['TSLA'])
adfuller_test(data['AMZN'])
adfuller_test(data['GOOGL'])
adfuller_test(data['JPM'])
adfuller_test(data['META'])
adfuller_test(data['NVDA'])
adfuller_test(data['TSM'])
```

```python
data['test_AAPL'] = data['AAPL']-data['AAPL'].shift(1)
data['test_LLY'] = data['LLY']-data['LLY'].shift(1)
data['test_MSFT'] = data['MSFT']-data['MSFT'].shift(1)
data['test_TSLA'] = data['TSLA']-data['TSLA'].shift(1)
data['test_AMZN'] = data['AMZN']-data['AMZN'].shift(1)
data['test_GOOGL'] = data['GOOGL']-data['GOOGL'].shift(1)
data['test_JPM'] = data['JPM']-data['JPM'].shift(1)
data['test_META'] = data['META']-data['META'].shift(1)
data['test_NVDA'] = data['NVDA']-data['NVDA'].shift(1)
data['test_TSM'] = data['TSM'] - data['TSM'].shift(1)
data
```

```python
adfuller_test(data['test_AAPL'].dropna())
adfuller_test(data['test_LLY'].dropna())
adfuller_test(data['test_MSFT'].dropna())
adfuller_test(data['test_TSLA'].dropna())
adfuller_test(data['test_AMZN'].dropna())
adfuller_test(data['test_GOOGL'].dropna())
adfuller_test(data['test_JPM'].dropna())
adfuller_test(data['test_META'].dropna())
adfuller_test(data['test_NVDA'].dropna())
adfuller_test(data['test_TSM'].dropna())
```

```python
import pandas as pd
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(data['NVDA'].dropna())
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(data['NVDA'].iloc[1:].dropna(),lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(data['NVDA'].iloc[1:].dropna(),lags=40,ax=ax2)

```

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(data['NVDA'],order=(2,1,2))
model_fit = model.fit()
data['forecast'] = model_fit.predict(start=2200,end=2500,dynamic=True)
data['display'] = data['NVDA'].iloc[1:]
data[['display','forecast']].plot(figsize=(12,8))
print(model_fit.summary())
```
